```python
import pandas as pd
import numpy as np
import timeit
from IPython.display import clear_output
```

```python
# dataframe of all chart instances for top40 songs from 1977 - 2022
long_df = pd.read_csv('Resources/top_40_cleaned_data_analysis.csv')

# dataframe with single instances for each top40 song, including audio and genre data
# some rows are missing data due to problems with Spotify API calls
short_df = pd.read_csv('../top40_audio_data_sample.csv')
```

```python
long_df['combo_str'] = long_df['artist'] + long_df['song']
short_df['combo_str'] = short_df['artist'] + short_df['song']
```

```python
long_df.tail()
```

```python
# rearrange short_df columns

all_columns = list(short_df.columns)

# select columns to rearrange
index_nums = [len(all_columns)-1] + list(range(5,len(all_columns)-1))
columns_to_paste = [all_columns[val] for val in index_nums]
new_short = short_df[['week','position','artist','song']]
short_df = pd.concat([new_short, short_df[columns_to_paste]], join='outer', axis=1)
short_df.head()
```

```python
columns_to_paste
```

```python
# restructure long_df to be formatted similarly to short_df
long_df.drop(columns=['Unnamed: 0'], inplace=True)
long_df = long_df[['week','position','artist','song','combo_str']]
```

```python
columns_to_paste = columns_to_paste[1:]
long_df2 = pd.DataFrame(columns=columns_to_paste)
long_df2.head()
```

```python
# add empty new columns to long data
long_df = pd.concat([long_df,long_df2])
#long_df[columns_to_paste] = np.nan
long_df.head()
```

```python
long_df.tail()
```

```python
short_df.tail()
```

```python
# check that columns are identical between both dfs
list(long_df.columns) == list(short_df.columns)
```

```python
long_df.reset_index(inplace=True)
short_df.reset_index(inplace=True)
```

```python
start = timeit.default_timer()
for l in long_df.index:
    for s in short_df.index:
    
        if long_df.at[l, 'combo_str'] == short_df.at[s, 'combo_str']:
            long_df.iloc[l] = short_df.iloc[s]

    # log progress
    if (l+1) % 100 == 0:
        clear_output()
        checkpoint = timeit.default_timer()
        print(f'Executed {l+1} of {len(long_df)} rows ({(l+1)/len(long_df)*100:.2f}% complete)')
        print(f'{(checkpoint-start)/60:.2f} minutes elapsed')

long_df.tail(10)
```

```python
long2 = long_df.dropna()
```

```python
long2.head()
```

```python
long3 = long2.drop_duplicates
```

```python
long2['combo'] = [i + j for i, j in zip(long2['artist_id'], long2['song_id'])]
```

```python
# rearrange columns for final output
final_df = pd.DataFrame(long2[['week','position','artist','song','combo','combo_str']])
final_df = pd.concat([final_df,long2.iloc[:,7:len(long2.columns)-1]],axis=1)
final_df['week'] = pd.to_datetime(final_df['week'])
final_df.sort_values(by=['week','position'], inplace=True)
final_df['chart_order'] = final_df.index
final_df = pd.concat([final_df['chart_order'], final_df.iloc[:,1:len(final_df.columns)]])
final_df.head(20)
```
