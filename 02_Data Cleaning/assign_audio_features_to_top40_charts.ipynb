```python
import pandas as pd
import numpy as np
import timeit
from IPython.display import clear_output
```

```python
# dataframe of all chart instances for top40 songs from 1977 - 2022
long_df = pd.read_csv('Resources/top_40_cleaned_data_analysis.csv')

# dataframe with single instances for each top40 song, including audio and genre data
# some rows are missing data due to problems with Spotify API calls
short_df = pd.read_csv('../top40_songs_audio.csv')
```

```python
long_df['combo_str'] = long_df['artist'] + ': ' + long_df['song']
short_df['combo_str'] = short_df['artist'] + ': ' +short_df['song']
```

```python
long_df.head()
```

```python
short_df.head()
```

```python
# join the dataframes: outer join to keep all chart instances
merged_df = long_df.merge(short_df, how='left', on='combo_str')
merged_df.head()
```

```python
list(merged_df.columns)
```

```python
# clean up the dataframe
merged_df.drop(columns=['Unnamed: 0', 'combo', 'week_y', 'position_y', 'artist_y', 'song_y'], inplace=True)
merged_df.rename(columns={'combo_str': 'combo', 'week_x': 'week', 'position_x': 'position', 'song_x': 'song', 'artist_x': 'artist'}, inplace=True)
merged_df.dropna(inplace=True)
merged_df.drop_duplicates(inplace=True)
print(merged_df.shape)
merged_df.head()
```

```python
# make sure that all instances of a song exist, have the same audio/genre data for different weeks/positions
merged_df.loc[merged_df.combo == 'Hanson: Mmm Bop'] # works as desired!
```

```python
# final df organization
merged_df['week'] = pd.to_datetime(merged_df['week'])
merged_df.sort_values(['week', 'position'], ascending=[True, True], ignore_index=True, inplace=True)
merged_df['chart_order'] = range(len(merged_df))
new_columns = [merged_df.columns[-1]] + list(merged_df.columns[:-1])

final_df = merged_df[new_columns]
final_df.head()
```

```python
# output the final merged df to csv
final_df.to_csv('../top40_charts_audio.csv', index=False)
```
