```python
from sqlalchemy import create_engine
import pandas as pd
import config
```

```python
# connection credentials for the 'top40' Amazon RDS Postgres database
db_name = 'top40'
db_url = 'top40.cext0azslr6v.us-west-1.rds.amazonaws.com'
db_username = 'postgres'
db_port = '5432'
db_password = config.db_password
```

```python
# connect to the database
# https://docs.sqlalchemy.org/en/14/core/engines.html#database-urls
# SQLalchemy connection format: dialect+driver://username:password@host:port/database

# create a connection string
db_string = f"postgresql://{db_username}:{db_password}@{db_url}/{db_name}"

# create an engine
engine = create_engine(db_string)
```

```python
# read in a dataframe
# tables in the database are 'songs', 'charts'

songs = pd.read_sql_query('select * from songs', con=engine)
songs.head()
```
